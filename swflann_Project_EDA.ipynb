```python
df4 = pd.read_csv('C:/Users/sflan/Dropbox (Brown)/Healthfully/Clean Data/weather_flu_google_9.csv')
```

```python
df4['cases'] = df4['percent.positive'] * df4['total.specimens']/100
df5 = df4.dropna()
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")

# Compute the correlation matrix
corr = df5.corr(method='spearman')

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr.sort_values(by='cases', axis=0, ascending=True).sort_values(by='cases', axis=1, ascending=False), mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
plt.scatter(df5['median_t'], df5['cases'])
plt.plot(np.unique(df5['median_t']), np.poly1d(np.polyfit(df5['median_t'], df5['cases'], 1))(np.unique(df5['median_t'])))
plt.xlabel('median temperature (degF)')
plt.ylabel('cases')
plt.title('Weekly Cases vs. Temperature')
```

```python
plt.scatter(df5['absolute_humidity'], df5['cases'])
plt.plot(np.unique(df5['absolute_humidity']), np.poly1d(np.polyfit(df5['absolute_humidity'], df5['cases'], 1))(np.unique(df5['absolute_humidity'])))
plt.xlabel('median humidity(%)')
plt.ylabel('cases')
plt.title('Weekly Cases vs. Absolute Humidity')
```

```python
plt.scatter(df5['tamiflu'], df5['cases'])
plt.plot(np.unique(df5['tamiflu']), np.poly1d(np.polyfit(df5['tamiflu'], df5['cases'], 1))(np.unique(df5['tamiflu'])))
plt.xlabel('tamiflu searches (%)')
plt.ylabel('cases')
plt.title('Weekly Cases vs. Tamiflu')
```

```python
plt.scatter(df5['flu_symptoms'], df5['cases'])
plt.plot(np.unique(df5['flu_symptoms']), np.poly1d(np.polyfit(df5['flu_symptoms'], df5['cases'], 1))(np.unique(df5['flu_symptoms'])))
plt.xlabel('flu symptoms searches (%)')
plt.ylabel('cases')
plt.title('Weekly Cases vs. Flu Symptoms')
```
