```python
import pandas as pd
weekly=pd.read_csv('full_weekly_data.csv')
monthly=pd.read_csv('full_monthly.csv')
```

```python
date=pd.DataFrame()
date[['month','day','year']]=weekly['week'].str.split('/',expand=True)
date['year']=date['year'].apply(lambda x: '20' +x)
date['month']=date['month'].apply(lambda x: x if len(x)==2 else '0'+x)
weekly['month']=date['year']+'-'+date['month']
```

```python
merged=weekly.merge(monthly,on=['month'])
final=pd.DataFrame()
#final['week_number']=(final.index+40)%52
#final.loc[169,'week_number']=53
#final.loc[170:,'week_number']=(final.loc[170:,'week_number']-1)%52
#final['week_number']=final['week_number'].replace(to_replace=0,value=52)
means=merged.groupby('month').mean()
merged=merged.merge(means,on=['month'])
for col in weekly.columns:
    if col != 'week' and col != 'month':
        final[col]=weekly[col]*merged[col+'_y_y']/(merged[col+'_x_y'].apply(lambda x: max(x,1)))
        final[col]=final[col].replace(to_replace=0,value=merged[col+'_y_y'])   
final.to_csv('google_data_cleaned.csv',index=False)
```
